```python
import pandas as pd

lst = pd.read_csv('https://raw.githubusercontent.com/KanoonGammy/Technical-Analysis-Project/refs/heads/main/SET_Index.csv')
lst_tickers = list(lst['symbol']+'.BK')
lst_tickers.append('^SET.BK')
print(lst_tickers)
```

```python
import yfinance as yf

# 1. กำหนดชื่อหุ้นที่ต้องการ (ต้องต่อท้ายด้วย .BK สำหรับหุ้นไทย)
ticker_symbol = lst_tickers

# 2. ดึงข้อมูลย้อนหลัง 1 ปี
# period สามารถเปลี่ยนได้ เช่น "1mo", "3mo", "5y", "max"
data = yf.download(ticker_symbol, period="10y")

# 3. แสดงผล 5 วันทำการล่าสุด
if not data.empty:
    print(f"ข้อมูลล่าสุดของหุ้น {ticker_symbol}:")
    print(data.tail())
else:
    print(f"ไม่สามารถดึงข้อมูลของหุ้น {ticker_symbol} ได้ อาจเป็นชื่อย่อที่ไม่ถูกต้อง")
```
